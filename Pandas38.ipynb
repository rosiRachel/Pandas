Mengenal parameter header dan skiprows
- Fungsi 'header', untuk menentukan header dari dataframe ada di baris berapa
- Ex: df = pd.read_csv('iris_error.csv', header=2)
- Fungsi 'skiprows', untuk menentukan baris berapa saja yang akan di lewat
- Ex: df = pd.read_csv('iris_error.csv', header=2, skiprows=[5,6])
