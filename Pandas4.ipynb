```python
a_rows = 10
a_cols = 1
cols1 = ('Usia',)
df = pd.DataFrame(np.random.randint(1, 99, size=(a_rows, a_cols)), columns=cols1)
df
```

```python
b_rows = 5
b_cols = 5
cols2 = tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1, 20, size=(b_rows, b_cols)), columns=cols2)
df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1, 4])
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0, 3])
df2
```

```python
df_inner = pd.merge(df1, df2, how='inner')
df_inner
```

```python
df_outer = pd.merge(df1, df2, how='outer')
df_outer
```
